```python
# Import des modules nécessaires
from collections import Counter
from typing import Dict
import tkinter as tk
import tkinter.scrolledtext as st
from collections import Counter

# Partie 1 : Structures abstraites de données


class Sommet:
    def __init__(self, étiquette):
        self.étiquette = étiquette
        self.gauche = None
        self.droite = None

    def modifier_étiquette(self, nouvelle_étiquette):
        self.étiquette = nouvelle_étiquette


class ArbreB:
    def __init__(self):
        self.racine = None

    def est_vide(self):
        return self.racine is None

    def inserer(self, étiquette):
        nouveau_sommet = Sommet(étiquette)
        if self.est_vide():
            self.racine = nouveau_sommet
        else:
            self._ajouter_sommet(nouveau_sommet, self.racine)

    def _ajouter_sommet(self, nouveau_sommet, sommet_courant):
        if nouveau_sommet.étiquette < sommet_courant.étiquette:
            if sommet_courant.gauche is None:
                sommet_courant.gauche = nouveau_sommet
            else:
                self._ajouter_sommet(nouveau_sommet, sommet_courant.gauche)
        else:
            if sommet_courant.droite is None:
                sommet_courant.droite = nouveau_sommet
            else:
                self._ajouter_sommet(nouveau_sommet, sommet_courant.droite)

     # Surcharge de l'opérateur d'insertion
    def __add__(self, etiquette):
        if self.etiquette is None:
            self.etiquette = etiquette
        elif etiquette < self.etiq:
            if self.fils_gauche is None:
                self.fils_gauche = ArbreB(etiquette)
            else:
                self.fils_gauche + etiquette
        else:
            if self.fils_droit is None:
                self.fils_droit = ArbreB(etiquette)
            else:
                self.fils_droit + etiquette
        return self

    def afficher(self):
        if self.racine is not None:
            self._afficher(self.racine)

    def _afficher(self, sommet_courant):
        if sommet_courant is not None:
            self._afficher(sommet_courant.gauche)
            print(sommet_courant.étiquette)
            self._afficher(sommet_courant.droite)

    def _rechercher(self, sommet_courant, étiquette):
        if sommet_courant is None:
            return None
        elif sommet_courant.étiquette == étiquette:
            return sommet_courant
        elif étiquette < sommet_courant.étiquette:
            return self._rechercher(sommet_courant.gauche, étiquette)
        else:
            return self._rechercher(sommet_courant.droite, étiquette)

    def rechercher(self, étiquette):
        if self.racine is None:
            return False
        else:
            noeud = self._rechercher(self.racine, étiquette)
            if noeud is None:
                return False
            else:
                return True
    # Surcharge de l'opérateur de recherche

    def __contains__(self, etiquette):
        if self.etiquette == etiquette:
            return True
        elif etiquette < self.etiquette and self.fils_gauche is not None:
            return etiquette in self.fils_gauche
        elif etiquette > self.etiquette and self.fils_droit is not None:
            return etiquette in self.fils_droit
        else:
            return False

    def fusion_arbres(self, arbre2):
        """
        Fusionner arbre  l'arbre courant avec l'arbre2 et retourne le résultat de la fusion
        """
        fusion = ArbreB()
        fusion.racine = self._fusionner(self.racine, arbre2.racine)
        return fusion

    def _fusionner(self, racine1, racine2):
        """
        Fonction récursive qui fusionne les sous-arbres ayant racine1 et racine2
        """
        if racine1 is None:
            return racine2
        if racine2 is None:
            return racine1

        # Création d'un nouveau sommet pour la fusion
        sommet_fusion = Sommet(racine1.étiquette + racine2.étiquette)

        # Fusion des sous-arbres gauche et droit
        sommet_fusion.gauche = self._fusionner(racine1.gauche, racine2.gauche)
        sommet_fusion.droite = self._fusionner(racine1.droite, racine2.droite)

        return sommet_fusion
    # Surcharge de l'opérateur d'addition pour la fusion d'arbres

    def __add__(self, other):
        if self.etiquette is None:
            return other
        elif other.etiquette is None:
            return self
        else:
            arbre_fusionne = ArbreB(self.etiquette)
            arbre_fusionne.fils_gauche = self.fils_gauche + other.fils_gauche
            arbre_fusionne.fils_droit = self.fils_droit + other.fils_droit

    def decomposer_arbres(self, étiquette):

        sous_arbre1 = ArbreB()
        sous_arbre2 = ArbreB()
        self._decomposer_sommets(
            self.racine, étiquette, sous_arbre1, sous_arbre2)
        return sous_arbre1, sous_arbre2

    def _decomposer_sommets(self, sommet_courant, étiquette, sous_arbre1, sous_arbre2):

        if sommet_courant is None:
            return

        if sommet_courant.étiquette < étiquette:
            sous_arbre1.inserer(sommet_courant.étiquette)
        else:
            sous_arbre2.inserer(sommet_courant.étiquette)

        self._decomposer_sommets(
            sommet_courant.gauche, étiquette, sous_arbre1, sous_arbre2)
        self._decomposer_sommets(
            sommet_courant.droite, étiquette, sous_arbre1, sous_arbre2)

    def supprimer(self, étiquette):
        # Cas où l'arbre est vide
        if self.racine is None:
            return self.racine

        # Recherche du sommet contenant la valeur à supprimer
        pere = None
        courant = self.racine
        while courant is not None and courant.étiquette != étiquette:
            pere = courant
            if étiquette < courant.étiquette:

                courant = courant.gauche
            else:
                courant = courant.droite

        # Cas où la valeur n'est pas présente dans l'arbre
        if courant is None:
            return self.racine

        # Cas où le sommet à supprimer n'a pas de fils
        if courant.gauche is None and courant.droite is None:
            if courant == self.racine:
                self.racine = None
            elif pere.gauche == courant:
                pere.gauche = None
            else:
                pere.droite = None

        # Cas où le sommet à supprimer a un seul fils
        elif courant.gauche is None:
            if courant == self.racine:
                self.racine = courant.droite
            elif pere.gauche == courant:
                pere.gauche = courant.droite
            else:
                pere.droite = courant.droite
        elif courant.droite is None:
            if courant == self.racine:
                self.racine = courant.gauche
            elif pere.gauche == courant:
                pere.gauche = courant.gauche
            else:
                pere.droite = courant.gauche
        # Cas où le sommet à supprimer a deux fils
        else:
            successeur = self._trouver_successeur(courant)
            if courant == self.racine:
                self.racine = successeur
            elif pere.gauche == courant:
                pere.gauche = successeur
            else:
                pere.droite = successeur

            successeur.gauche = courant.gauche

        return self.racine

    def _trouver_successeur(self, sommet_courant):
        successeur_pere = sommet_courant
        successeur = sommet_courant
        courant = sommet_courant.droite

        while courant is not None:
            successeur_pere = successeur
            successeur = courant
            courant = courant.gauche

        if successeur != sommet_courant.droite:
            successeur_pere.gauche = successeur.droite
            successeur.droite = sommet_courant.droite

        return successeur
     # Surcharge de l'opérateur de suppression

    def __sub__(self, etiquette):
        if self.etiq is None:
            return self
        if etiquette < self.etiquette:
            if self.fils_gauche is not None:
                self.fils_gauche = self.fils_gauche - etiquette
            return self
        elif etiquette > self.etiquette:
            if self.fils_droit is not None:
                self.fils_droit = self.fils_droit - etiquette
            return self
        else:
            if self.fils_gauche is None:
                return self.fils_droit
            elif self.fils_droit is None:
                return self.fils_gauche
            else:
                temp = self.fils_droit
                while temp.fils_gauche is not None:
                    temp = temp.fils_gauche
                self.etiquette = temp.etiquette
                self.fils_droit = self.fils_droit - temp.etiquette
            return self


class VisualiseurArbre:
    def __init__(self, arbre):
        self.arbre = arbre
        self.fenetre = tk.Tk()
        self.canvas = tk.Canvas(self.fenetre, width=500, height=500)
        self.canvas.pack()

        self.dessiner_arbre(self.arbre.racine, 250, 50, 200)

        self.fenetre.mainloop()

    def dessiner_arbre(self, noeud, x, y, espace):
        if noeud:
            self.canvas.create_oval(
                x-15, y-15, x+15, y+15, fill="white", width=2)
            self.canvas.create_text(x, y, text=noeud.étiquette)

            espace = espace // 2 if espace > 50 else 50

            if noeud.gauche:
                self.canvas.create_line(x, y, x-espace, y+50, width=2)
                self.dessiner_arbre(noeud.gauche, x-espace, y+50, espace)

            if noeud.droite:
                self.canvas.create_line(x, y, x+espace, y+50, width=2)
                self.dessiner_arbre(noeud.droite, x+espace, y+50, espace)


# Partie 2 : Cryptage

# Définition de la classe ArbreHuffman


class ArbreHuffman(ArbreB):
    def __init__(self, freq: float, caractere: str = None, gauche: 'ArbreHuffman' = None, droite: 'ArbreHuffman' = None):
        super().__init__()
        self.freq = freq
        self.étiquette = caractere
        self.gauche = gauche
        self.droite = droite

    def est_feuille(self) -> bool:
        return not self.gauche and not self.droite

    def __lt__(self, other: 'ArbreHuffman') -> bool:
        return self.freq < other.freq

# Fonction pour construire l'arbre de Huffman


def huffman_encoding(data):
    frequences = dict(Counter(data))
    total_caracteres = sum(frequences.values())
    pourcentage_occurrences = {
        char: (count / total_caracteres) * 100 for char, count in frequences.items()}
    sommets = [ArbreHuffman(pourcentage_occurrences[char], char)
               for char in pourcentage_occurrences]
    while len(sommets) > 1:
        sommets = sorted(sommets, key=lambda node: node.freq)
        gauche = sommets.pop(0)
        droite = sommets.pop(0)
        parent = ArbreHuffman(gauche.freq + droite.freq)
        parent.gauche = gauche
        parent.droite = droite
        sommets.append(parent)
    root = sommets[0]
    stack = [(root, "")]
    while stack:
        node, code = stack.pop()
        node.code = code
        if node.gauche:
            stack.append((node.gauche, code + "0"))
        if node.droite:
            stack.append((node.droite, code + "1"))
    return root


# Construction de l'arbre binaire de cryptage
# On a choisi ce texte car c'est un pangramme : il contient toutes les lettres de l'alphabet. On a ajouté à la fin une apostrophe
# et un point afin de pouvoir construire des phrases et des contractions. 
arbre = huffman_encoding("portez ce vieux whisky au juge blond qui fume '.")


# Affichage de l'arbre binaire de cryptage dans le terminal de commande


def afficher_arbre(noeud: ArbreHuffman, prefixe: str = '', est_gauche: bool = False):
    if noeud:
        print(prefixe + ('└──' if est_gauche else '┌──') +
              str(noeud.étiquette) + " (" + str(noeud.freq) + ")")
        prefixe += '    ' if est_gauche else '│   '
        afficher_arbre(noeud.gauche, prefixe, True)
        afficher_arbre(noeud.droite, prefixe, False)


# Fonction pour créer le dictionnaire de correspondance entre chaque caractère et son code binaire


def construire_dictionnaire_encodage(noeud: ArbreHuffman, code: str = '') -> Dict[str, str]:
    if not noeud:
        return {}

    if noeud.est_feuille():
        return {noeud.étiquette: code}

    dico_gauche = construire_dictionnaire_encodage(noeud.gauche, code + '0')
    dico_droite = construire_dictionnaire_encodage(noeud.droite, code + '1')

    return {**dico_gauche, **dico_droite}


# Création du dictionnaire de correspondance entre chaque caractère et son code
dico_codes_binaires = construire_dictionnaire_encodage(arbre)

# Utilisation du dictionnaire de correspondance pour encoder le texte en binaire


def encoder_texte(texte):
    texte_encode = ''.join(
        dico_codes_binaires[caractere] for caractere in texte)

    return texte_encode


# Partie 3 : Décryptage


def decoder_texte(texte_encode: str, arbre: ArbreHuffman) -> str:
    # Le code de décodage que nous avons discuté précédemment
    texte_decode = ""
    noeud_courant = arbre
    for bit in texte_encode:
        if bit == "0":
            noeud_courant = noeud_courant.gauche
        else:
            noeud_courant = noeud_courant.droite

        if noeud_courant.est_feuille():
            texte_decode += noeud_courant.étiquette
            noeud_courant = arbre

    return texte_decode


# Fonctions de l'interface graphique

# Fonction pour gérer le clic sur le bouton de conversion

def cliquer_pour_encoder():
    texte_original.delete(1.0, tk.END)
    texte_code.delete(1.0, tk.END)
    texte = texte_a_encoder.get()
    texte_encode = encoder_texte(texte)
    texte_original.insert(1.0, texte)
    texte_code.insert(1.0, texte_encode)


# Fonction pour décoder le texte lorsque le bouton "Décoder" est cliqué


def cliquer_pour_decoder():
    texte_original.delete(1.0, tk.END)
    texte_code.delete(1.0, tk.END)
    # obtenir le texte encodé à partir de la zone de texte d'entrée
    texte_encode = texte_a_encoder.get()
    texte_decode = decoder_texte(texte_encode, arbre)  # décoder le texte
    # afficher le texte décodé sur l'étiquette de résultat
    texte_code.insert(1.0, texte_decode)
    texte_original.insert(1.0, texte_encode)


def draw_tree(canvas, node, x, y, x_offset, y_offset):
    if node is not None:
        if node.étiquette is not None:
            text = node.étiquette + " - " + str(node.freq) + " - " + node.code
        else:
            text = " - " + str(node.freq) + " - " + node.code
        canvas.create_text(x, y, text=text)
        if node.gauche is not None:
            canvas.create_line(x, y, x - x_offset, y + y_offset)
            draw_tree(canvas, node.gauche, x - x_offset,
                      y + y_offset, x_offset/2, y_offset)
        if node.droite is not None:
            canvas.create_line(x, y, x + x_offset, y + y_offset)
            draw_tree(canvas, node.droite, x + x_offset,
                      y + y_offset, x_offset/2, y_offset)


# Interface graphique

# Créer la fenêtre principale
fenetre = tk.Tk()
fenetre.geometry("1000x700")
fenetre.title("Encodeur et décodeur de texte")

# Ajouter les éléments de l'interface
texte_a_encoder_label = tk.Label(fenetre, text="Entrez le texte à encoder en minuscule et sans accent. Les caractères spéciaux possibles sont l'apostrophe et le point. : ")
texte_a_encoder = tk.Entry(fenetre)
texte_a_encoder_label.grid(row=0, column=0)
texte_a_encoder.grid(row=1, column=0)

bouton_encoder = tk.Button(fenetre, text="Encoder",
                           command=cliquer_pour_encoder)
bouton_encoder.grid(row=2, column=0)
bouton_decoder = tk.Button(fenetre, text="Décoder",
                           command=cliquer_pour_decoder)
bouton_decoder.grid(row=3, column=0)

texte_original = st.ScrolledText(
    fenetre, width=30, height=8, font=("Times New Roman", 15))
texte_code = st.ScrolledText(
    fenetre,  width=30, height=8, font=("Times New Roman", 15))
texte_original.insert(1.0, "texte original")
texte_code.insert(1.0, "texte encrypté ou décrypté")
texte_original.grid(row=4, column=0, columnspan=2)
texte_code.grid(row=5, column=0, columnspan=2)

canvas = tk.Canvas(fenetre, width=400, height=600)
scrollbar1 = tk.Scrollbar(fenetre, orient='vertical', command=canvas.yview)
scrollbar2 = tk.Scrollbar(fenetre, orient='horizontal', command=canvas.xview)
scrollbar1.grid(row=0, column=2)
scrollbar2.grid(row=1, column=2)
canvas.grid(row=2, column=2, rowspan=6)

text = "portez ce vieux whisky au juge blond qui fume '."
root = huffman_encoding(text)
draw_tree(canvas, root, 2700, 50, 2700, 50)


# Lancer la boucle principale
fenetre.mainloop()
```
